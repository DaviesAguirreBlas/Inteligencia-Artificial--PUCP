```
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```
#importamos el dataset

iris = datasets.load_iris()
iris.head()
```

```
#Vamos a dividir los datos

#Por un lado usaremos la información de iris para entrenar nuestro algoritmo
X_iris = iris.data

#Por otro lado le daremos la etiqueta, pero no vamos a entrenarlo a partir de esta etiqueta , sino con las similitudes que este dataset tenga

Y_iris = iris.target
```

```
#Crearemos nuestro "dataframe"

#X nos permitirá hacer el aprendizaje
x = pd.DataFrame(iris.data, columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
#Ahora pasaremos a y la información de los targets
y = pd.DataFrame(iris.target, columns = ['Target'])
x.head()
```

```
#Pasaremos a visualizar los puntos

plt.scatter(x['Petal Length'], x['Petal Width'], c = 'blue')
plt.xlabel('Petal Length', fontsize = 10)
plt.ylabel('Petal Width', fontsize = 10)
```

```
#Imprimiendo similitudes con el sépalo
#plt.sca
print(x)
type(x)
```

Modelado y Entrenamiento de nuestro algoritmo


```
#Empezamos a crear el modelo de "KMeans", le ingresaremos 2 parámetros

model  = KMeans(n_clusters = 2 , max_iter = 1000 )  # "n_cluster" indica la cantidad de K que queremos que tenga, la cantidad de entroides en nuestra gráfica o plano.
                                   #El otro parámetro es el número de iteraciones


#AHORA PASAMOS A ENTRENARLO

model.fit(x)  # "x" son nuestros puntos y con ello lo estaremos entrenando

#Todas aquellas etiquetas que encontró basado en los datos que encontró
y_labels = model.labels_
```

```
#Vamos a ver cómo ha dividido la información

y_kmeans = model.predict(x) #Hacemos la predicción con predict
print('Predicciones', y_kmeans)
```

```
#Veremos el accuracy para verificar la precisión del algorito para ello importamos el método " metrics"

from sklearn import metrics
# Recuerda esto, ya que es un parámetro que usaremos para el accuracy " Y_iris = iris.target"
accuracy = metrics.adjusted_rand_score(Y_iris, y_kmeans) #Con esto calclulamos el accuracy.Recuerda que le entregamos las Y_iris y nuestra predicción para esos valores
print(accuracy)

```
