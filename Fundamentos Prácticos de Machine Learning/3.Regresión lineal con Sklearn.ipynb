```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```
#Cargamos nuestro dataset con pandas.Recuerda, aquí en collab debes de subir el archivo a "files" para que pueda usarla.
dataset = pd.read_csv('salarios.csv')
dataset.head(5)
```

```
#Ahora veremos cuántos elementos tenemos en el dataset
dataset.shape
```

```
#La instrucción ".iloc" es para localiar cuáles son los datos que debemos utiliar para entrenar
#le estamos dando la columna 
x = dataset.iloc[:, :-1].values
#La variable "y" nos va a ayudar a predecir
y = dataset.iloc[:, 1].values
```

```
#Ahora pasaremos a entrenar a el algoritmo
#Dividiremos los datos con el modelo de Sklearn "train_test_split"

# A este método le daremos (x, y) que son nuestros datos que nos permiten poder partir la informcación para poder darle el feature y el label
#Luego, le pasamos "test_size" que es el porcentaje del total de los datos para testear el algoritmo
# "Random_state =0" lo usamos para evitar que cuando habramos este notebook de nuevo obtengamos otros datos, recuerda la aleatoriedad .Con este método evitamos otra aleatoriedad futura.

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size =0.2, random_state = 0)

```

```
Y_train
```

```
#creamos una variable para asignarle el modelo de regresion lineal

regressor =LinearRegression()

#La siguiente línea entrenará nuestro algoritmo ".fit()" es la clave para todo!
regressor.fit(X_train, Y_train)

#Con lo anterior ya hemos creado nuestro modelo y entrenado.

#
```

```
print(X_train)
print(Y_train)
```

**Entonces, recuerda** primero toma los datos; divide los datos para entrenamiento y prueba; y, luego, entrena con la instrucción .fit()  y de ahí en adelante hará el aprendizaje.


```
#Mostraremos lo que aprendió


#Aquí instanciamos a plt y le decimos a viz_train que sea un gráfico
viz_train = plt
#Con ".scatter" muestraremos nuestros elementos
viz_train.scatter(X_train, Y_train, color = 'blue')
#".predict()" nos mostrará nuestro regressor: Nuestra predicción. Ello mostrará nuestra linea cuyos valores son todos nuestros valores que estamos prediciendo
viz_train.plot(X_train, regressor.predict(X_train), color = 'black')
#Mostraremos los títulos para que sea entendible nuestro gráfico
viz_train.title('Salario Vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```
#Ahora vamos a validarlo de nuevo, pero con nuestros datos de prueba
viz_train = plt
viz_train.scatter(X_test, Y_test, color = 'red')
viz_train.plot(X_train, regressor.predict(X_train), color = 'black')
viz_train.title('Salario Vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```
regressor.score(X_test, Y_test)
```
