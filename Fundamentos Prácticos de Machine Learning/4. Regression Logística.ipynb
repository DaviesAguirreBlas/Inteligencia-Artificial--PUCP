```
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
diabetes = pd.read_csv('diabetes.csv')
diabetes.head(5)
```

```
#Definimos las columnas que vamos a utilizar, las que son necesarias para hacer el aprendizaje

feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Glucose', 'BloodPressure','DiabetesPedigreeFunction']

#Ahora llamaremos a nuestros target, los aributos con los que vamos a aprender
x = diabetes[feature_cols]
# "y" nos permitirá saber si la persona tiene diabetes o no de acuerdo a los datos anteriores
y = diabetes.Outcome
```

```
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.25, random_state =0 )
```

```
#Ahora llamaremos a nuestro algoritmo, nuestro modelo

logreg = LogisticRegression()

logreg.fit(X_train, Y_train)

y_pred = logreg.predict(X_test)
```

```
y_pred
```

```
import numpy as np

#Creamos nuestra matriz de difusion
cnf_matrix = metrics.confusion_matrix(Y_test, y_pred)

#Definimos nuestros las clases de datos tendrá
#Recuerda que '0' es cuando no tiene diabetes y '1' es cuando sí lo tiene
class_names = [0,1]
#creamos una figura para poder plotear
fig, ax = plt.subplots()

#Con la siguiente línea es para que nos muestre información en nuestra gráfica
tick_marks =np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues_r', fmt = 'g')
ax.xaxis.set_label_position('bottom')
plt.tight_layout()
plt.title('Matriz de confusion', y = 1.1)
plt.ylabel('Etiqueta actual ')
plt.xlabel('Etiqueta de prediccion')
```
